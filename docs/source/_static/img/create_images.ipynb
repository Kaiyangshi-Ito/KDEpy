# Minimal working example

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
```

```python
from KDEpy import KDE
import numpy as np
from scipy.stats import norm
data = norm(loc=0, scale=1).rvs(100)
x = np.linspace(-3, 3)
y = KDE(kernel='gaussian', bw=0.5).fit(data).evaluate(x)
```

```python
import numpy as np
plt.figure(figsize=(6, 3))
plt.plot(x, y, label='KDE estimate')
plt.plot(x, norm(loc=0, scale=1).pdf(x), label='True PDF')
plt.scatter(data, np.zeros_like(data), marker='+', color='red')
plt.legend(loc='best')
plt.tight_layout()
plt.savefig('minimal_working_example.png', dpi=200)
```

# Logo

```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)

fig, ax = plt.subplots(figsize=(5, 3))
x = np.linspace(-4, 4, 500)
#plt.plot(x, norm(loc=0, scale=1).pdf(x))
ax.fill_between(x, norm(loc=0, scale=1).pdf(x), zorder=15, alpha=1)
data = norm(loc=0, scale=1).rvs(10)
ax.scatter(data, np.zeros_like(data), marker='x', color='red', s=50, zorder=25)
ax.grid(True, alpha=0.5, zorder=-50)
ax.set_xticks(np.linspace(-3, 3, num=5))
ax.set_xticklabels([])
ax.set_yticks(np.linspace(0, np.max(norm(loc=0, scale=1).pdf(x)), num=5))
ax.set_yticklabels([])
ax.text(0.5, 0.5, r'$\hat{f}(x) = \frac{1}{h} \sum_{i=1}^n w_i K \left (  \frac{x-x_i}{h} \right )$', 
        horizontalalignment='center', verticalalignment='center', transform=ax.transAxes, zorder= 35, fontsize=20)

# add an ellipse
import matplotlib.patches as mpatches
ellipse = mpatches.Ellipse([0.5, 0.5], 0.95, 0.4, transform=ax.transAxes, color='green', zorder = 30, alpha=0.1)
ax.add_patch(ellipse)

plt.tight_layout()
plt.savefig('logo.png', dpi=200)
```
