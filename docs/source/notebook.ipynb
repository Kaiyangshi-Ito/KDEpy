```python
plt.figure()

x = np.linspace(-3, 3, num = 500)

for name, function in KDE._available_kernels.items():
    y = function(x)
    plt.plot(x, y, label=name.capitalize())
    plt.ylim(0, 0.5)
    
plt.legend(loc='best')
```
